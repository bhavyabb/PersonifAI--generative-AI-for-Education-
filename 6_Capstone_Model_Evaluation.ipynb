```python
from datasets import load_dataset
dataset_name = "aumghag/Data-Analytics-Digital-Marketing-Project-Management-QA_DB"
dataset = load_dataset(dataset_name)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load your fine-tuned model and tokenizer
model_name = "aumghag/ft-llama-3.1-8B-capstone-personifAI-edu"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
import torch

# Select a batch of examples (assuming dataset is structured in QA pairs)
questions = dataset['train']['question'][:10]  # Adjust the size accordingly
answers = dataset['train']['answer'][:10]
```

```python
predictions = []
for question in questions:
    inputs = tokenizer(question, return_tensors="pt").input_ids
    outputs = model.generate(inputs, max_length=50)
    predicted_answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    predictions.append(predicted_answer)
```

```python
from bert_score import score

# Compute BERTScore for predictions
P, R, F1 = score(predictions, answers, lang="en", verbose=True)

print(f"Average BERT Precision: {P.mean().item()*100}")
print(f"Average BERT Recall: {R.mean().item()*100}")
print(f"Average BERT F1 Score: {F1.mean().item()*100}")
```
